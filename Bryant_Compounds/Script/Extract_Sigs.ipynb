Script to extract signature IDs for extracting from LINCS

```python
import pandas as pd
```

```python
bryant_file = pd.read_csv("../Metadata/bestguessofcell-time-conc-touse.csv")
bryant_file
```

```python
# Distribution of cell stress types
df = bryant_file
df.groupby(['samplecode']).agg(['count'])
```

```python
# Import LINCS metadata
lincs_meta_1 = pd.read_csv("../../Coverage_Investigation/HEPG2_only/LINCS_Metadata/GSE70138_Broad_LINCS_sig_info_2017-03-06.txt",sep="\t")
lincs_meta_2 = pd.read_csv("../../Coverage_Investigation/HEPG2_only/LINCS_Metadata/GSE92742_Broad_LINCS_sig_info.txt",sep="\t")
```

```python
print(lincs_meta_1.columns)
print(lincs_meta_2.columns)
```

```python
# First subset meta 1
df.columns
all_codes = df['samplecode'].tolist()
pert_inames = df['pert_iname'].tolist()
cell_types = df['cell_type'].tolist()

meta_1_subset = lincs_meta_1[(lincs_meta_1['pert_iname'].isin(pert_inames))&(lincs_meta_1['cell_id'].isin(cell_types))]
meta_1_subset
```

```python
# Get list of sig_id
sig_ids_to_extract = meta_1_subset['sig_id'].tolist()
sig_ids_to_extract
```

```python
# Now repeat for phase 2
meta_2_subset = lincs_meta_2[(lincs_meta_2['pert_iname'].isin(pert_inames))&(lincs_meta_2['cell_id'].isin(cell_types))]
meta_2_subset
```

```python
sig_ids_to_extract_2 = meta_2_subset['sig_id'].tolist()
sig_ids_to_extract_2
```

```python
all_sig_ids = sig_ids_to_extract + sig_ids_to_extract_2
all_sig_ids
```

```python
# Save as .txt file
all_sig_ids = pd.DataFrame(all_sig_ids)
all_sig_ids.to_csv("../Metadata/sig_ids_all.txt",sep="\t")
```
