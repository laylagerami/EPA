```python
# Extract data coverage in HepG2 in LINCS
import pandas as pd

compound_anno = pd.read_csv("../Compound_Annotations/compounds_and_classes.csv")
compound_anno
```

```python
meta_1 = pd.read_csv("../LINCS_Metadata/GSE92742_Broad_LINCS_sig_info.txt",sep="\t")
meta_2 = pd.read_csv("../LINCS_Metadata/GSE70138_Broad_LINCS_sig_info_2017-03-06.txt",sep="\t")
```

```python
meta_1.head()
```

```python
print(meta_1.columns.tolist())
```

```python
meta_2.head()
```

```python
print(meta_2.columns.tolist())
```

```python
all_perts = compound_anno['pert_iname'].tolist()
all_perts[0:5]
```

```python
meta_1_perts = meta_1[(meta_1['pert_iname'].isin(all_perts)) & (meta_1['cell_id']=="HEPG2")]
meta_1_perts
```

```python
meta_2_perts = meta_2[(meta_2['pert_iname'].isin(all_perts)) & (meta_2['cell_id']=="HEPG2")]
meta_2_perts
```

```python
# subset each df
meta_1_df = meta_1_perts[['sig_id','pert_iname','pert_idose','pert_itime','distil_id']]
meta_1_df
```

```python
meta_2_df = meta_2_perts[['sig_id','pert_iname','pert_idose','pert_itime','distil_id']]
meta_2_df
```

```python
# merge
all_meta = pd.concat([meta_1_df,meta_2_df])
all_meta.head()
```

```python
# add the class annotation
compound_anno = compound_anno[['PMI guided stresscat','pert_iname']]
compound_anno
```

```python
all_meta_and_anno = all_meta.merge(compound_anno,on="pert_iname")
```

```python
all_meta_and_anno
```

```python
all_meta_and_anno.to_csv("../Results/HEPG2_data.csv")
```
