### Script to convert sig_ids to compound + conditions, plus collapse replicates

```python
# Import pandas
import pandas as pd
```

```python
# Read in data matrix
lincs_data = pd.read_csv("extracted_signatures_epa.txt",sep="\t")
lincs_data.head()
```

```python
# Read in sig_id to conditions file
sig_ids_1 = pd.read_csv("sig_ids_1.csv")
sig_ids_2 = pd.read_csv("sig_ids_2.csv")
```

```python
# Concatenate compound_condition for mapping (1)
sig_ids_1
```

```python
sig_ids_1['new_colname'] = sig_ids_1['pert_iname'].astype(str) + '_' + sig_ids_1['cell_id'] + '_' + sig_ids_1['pert_itime'] + '_' + sig_ids_1['pert_idose']
sig_ids_1_new = sig_ids_1[['sig_id','new_colname']]
sig_ids_1_new
```

```python
# DO the same for sig_ids_2
sig_ids_2
```

```python
sig_ids_2['new_colname'] = sig_ids_2['pert_iname'].astype(str) + '_' + sig_ids_2['cell_id'] + '_' + sig_ids_2['pert_itime'] + '_' + sig_ids_2['pert_idose']
sig_ids_2_new = sig_ids_2[['sig_id','new_colname']]
sig_ids_2_new
```

```python
# Stick them together
full_mapping = pd.concat([sig_ids_1_new,sig_ids_2_new])
full_mapping
```

```python
# Now change the Sig_id in the original df
mapping = dict(full_mapping[['sig_id', 'new_colname']].values)
lincs_data['Experiment'] = lincs_data.Sig_id.map(mapping)
lincs_data
```

```python
# Collapsing replicates
lincs_data.groupby(['Experiment']).count().sort_values(['Sig_id'], ascending=False)
```

```python
# Loop
sigs = []
import numpy as np
for exp in list(set(lincs_data['Experiment'].tolist())):
    subset = lincs_data[lincs_data['Experiment']==exp]
    subset_t = subset.T
    new_header = subset_t.iloc[-1] #grab the last row for the header
    subset_t = subset_t[1:] #take the data less the header row
    subset_t.drop(subset_t.tail(1).index,inplace=True)
    subset_t.columns = new_header #set the header row as the df header
    subset_t = subset_t.apply(pd.to_numeric, errors='coerce')
    if subset.shape[0] > 1:
        corr_reps = subset_t.corr()
        a = np.array(corr_reps)
        corr_reps_nan = pd.DataFrame(a)
        col_weights = []
        for col in corr_reps_nan:
            col_weight = corr_reps_nan[col].sum()
            col_weights.append(col_weight)
        col_weights_norm = [float(i)/sum(col_weights) for i in col_weights]
        linear_list = []
        for idx, weight in enumerate(col_weights_norm):
            linear = subset_t.iloc[:,idx]*col_weights_norm[idx]
            linear_list.append(linear)
        combined_sig = sum(linear_list)
        combined_sig = list(map(str, combined_sig))
        consensus_sig = combined_sig
        consensus_sig.insert(0,exp)
        consensus_sig = pd.DataFrame(consensus_sig)
        sigs.append(consensus_sig)
    else:
        subset_t = pd.DataFrame([exp]+subset_t.iloc[:,0].tolist())
        sigs.append(subset_t)
        
```

```python
# Tidying
df = pd.concat(sigs,axis=1)
header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = header #set the header row as the df header
df['Gene_Id'] = lincs_data.columns.tolist()[1:-1]
col = df.pop("Gene_Id")
df.insert(0, col.name, col)
df
```

```python
df.to_csv("final_lincs_data.csv")
```
