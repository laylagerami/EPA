## Get Sig IDs for subset of compounds

```python
# import packages
import pandas as pd
```

```python
# read in metadata (phase 1 and 2) and Janssen compound InChI keys
meta1 = pd.read_csv("../../JanssenProject/LINCS/Input/LINCS_Metadata/sig_info/GSE70138_Broad_LINCS_sig_info_2017-03-06.txt",sep="\t")
meta2 = pd.read_csv("../../JanssenProject/LINCS/Input/LINCS_Metadata/sig_info/GSE92742_Broad_LINCS_sig_info.txt",sep="\t")
bryant_comps = pd.read_csv("compound_list.csv")
```

```python
# Extract pert info (LINCS compound inchi keys)
pert_1 = pd.read_csv("../../JanssenProject/LINCS/Input/LINCS_Metadata/pert_info/GSE70138_Broad_LINCS_pert_info (1).txt",sep="\t")
pert_2 = pd.read_csv("../../JanssenProject/LINCS/Input/LINCS_Metadata/pert_info/GSE92742_Broad_LINCS_pert_info.txt",sep="\t")
```

```python
# Explore pert info
pert_1
```

```python
# Extract comp names needed
bryant_comps = bryant_comps['name'].tolist()
```

```python
# List comps
bryant_comps
```

```python
# Now get exp info from metadata (i.e. sig_ids)
exp1 = meta1[meta1['pert_iname'].isin(bryant_comps)]
exp2 = meta2[meta2['pert_iname'].isin(bryant_comps)]
```

```python
# Filter for cells
cells = ['JURKAT','HEPG2','HCC515','VCAP','MCF4']
exp1 = exp1[exp1['cell_id'].isin(cells)]
exp2 = exp2[exp2['cell_id'].isin(cells)]
```
